# Extract Data From sqlite-Database

After conducting the web crawl, all logged browser events are stored into an sqlite database. For the further steps of the analysis, we had to extract the required information from the database.

## HTTP Request Data

```python
#read data from sqlite database into pandas dataframe
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('C:/Users/timpe_000/Desktop/crawl-data.sqlite')

requests_df = pd.read_sql_query("SELECT visit_id, url, top_level_url FROM http_requests", conn)
```

```python
#save as csv
requests_df.to_csv("http_requests.csv", index=False)
```

## JavaScript Calls Data

```python
#read data from sqlite database into pandas dataframe
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('C:/Users/timpe_000/Desktop/crawl-data.sqlite')

js_df = pd.read_sql_query("SELECT script_url, symbol, operation, value, top_level_url, func_name from javascript", conn)
```

```python
js_df.columns = ['script_urls', 'symbol', 'operation', 'value', 'top_level_url', 'func_name']

js_df.script_urls[:10]
```

```python
#save as csv
js_df.to_csv("javascript.csv", index=False)
```

```python
import pandas as pd

#load data
http_requests_df = pd.read_csv("/content/drive/My Drive/IT_Security/Code and Data/http_requests.csv")

http_requests_df
```

```python
http_requests_df['url_len'] = http_requests_df.url.apply(lambda x: len(x))
http_requests_df['top_level_url_len'] = http_requests_df.top_level_url.apply(lambda x: len(x))

selected_http_requests_df = http_requests_df.loc[(http_requests_df['url_len'] <= 25) & (http_requests_df.url != http_requests_df.top_level_url) & (http_requests_df.top_level_url_len < 25)]
selected_http_requests_df = selected_http_requests_df[['visit_id', 'url', 'top_level_url']].iloc[[0,1,2,3,10,12,13,14,15,20]]
selected_http_requests_df.to_csv("selected_http_requests_df.csv", index=False)
selected_http_requests_df
```

```python
import pandas as pd

#load javascript data
javascript_df = pd.read_csv("/content/drive/My Drive/IT_Security/Code and Data/javascript.csv")

javascript_df
```

```python
javascript_df['script_url'] = javascript_df.script_url.apply(lambda x: str(x))
javascript_df['script_url_len'] = javascript_df.script_url.apply(lambda x: len(x))
javascript_df['top_level_url_len'] = javascript_df.top_level_url.apply(lambda x: len(x))

javascript_df

selected_javascript_df1 = javascript_df.loc[(javascript_df['script_url_len'] <= 30) & (javascript_df.top_level_url_len < 20) & (javascript_df.value.isnull() == False)]
selected_javascript_df1 = selected_javascript_df1[['visit_id', 'script_url', 'symbol', 'operation', 'value', 'arguments', 'top_level_url']]

selected_javascript_df2 = javascript_df.loc[(javascript_df['script_url_len'] <= 30) & (javascript_df.top_level_url_len < 20) & (javascript_df.arguments.isnull() == False)]
selected_javascript_df2 = selected_javascript_df2[['visit_id', 'script_url', 'symbol', 'operation', 'value', 'arguments', 'top_level_url']]

selected_javascript_df1 = selected_javascript_df1.iloc[[22, 40, 50, 60, 111], :]
selected_javascript_df2 = selected_javascript_df2.iloc[[999, 5, 13, 199, 202], :]

selected_javascript_df = pd.concat([selected_javascript_df1, selected_javascript_df2])
selected_javascript_df.to_csv("selected_javascript_df.csv", index=False)
selected_javascript_df
```
