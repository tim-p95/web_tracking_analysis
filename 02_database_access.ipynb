# Extract Data From sqlite-Database

After conducting the web crawl, all logged browser events are stored into an sqlite database. For the further steps of the analysis, we had to extract the required information from the database.

## HTTP Request Data

```python
#read data from sqlite database into pandas dataframe
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('C:/Users/timpe_000/Desktop/crawl-data.sqlite')

requests_df = pd.read_sql_query("SELECT visit_id, url, top_level_url FROM http_requests", conn)
```

```python
#save as csv
requests_df.to_csv("http_requests.csv", index=False)
```

## JavaScript Calls Data

```python
#read data from sqlite database into pandas dataframe
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('C:/Users/timpe_000/Desktop/crawl-data.sqlite')

js_df = pd.read_sql_query("SELECT script_url, symbol, operation, value, top_level_url, func_name from javascript", conn)
```

```python
js_df.columns = ['script_urls', 'symbol', 'operation', 'value', 'top_level_url', 'func_name']

js_df.script_urls[:10]
```

```python
#save as csv
js_df.to_csv("javascript.csv", index=False)
```
